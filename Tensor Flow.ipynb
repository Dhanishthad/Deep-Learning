```python
#import lib
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('C:/Users/Dhanishtha/OneDrive/Documents/Placement Preparation/Data Science/TensorFlow/Churn.csv')
```

```python
X=pd.get_dummies(df.drop(['Churn','Customer ID'],axis=1 ))
y=df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=2)
```

```python
y_train.head()
```

1. Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model #allows to reload our model from memoery later on 
from tensorflow.keras.layers import Dense #Dense is going to be our fully connected layer in our neural network which allows us to build a whole bunch of hidden layersk
#Types of Layers: you can also import layers optimized for images and natural languages as well e.g. Conv2D and LSTM
from sklearn.metrics import accuracy_score 
```

2. Build and Compile our model

```python
#We will add dense layers
model= Sequential() 
#add more layers by using model.add() through keep in mind having more layers will take longer to train
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
#units=32 means we are going to have 32 neurons inside this dense layer
#the activation function acts as our modifier for our output
#using relu we are converting the output to a minimum of zero and then an unlimited upward value
#input_dim=len(X_train.columns) it is gonna be the same no. of dimensions that are available inside our feature data framee
model.add(Dense(units=64, activation='relu'))
#Final layer shape: the no. of units in our last layer dictates what our output will look like,
#one unit will mean that we only get onw no. back, sophisticated models for other tasks often have multiple units in the final layer 
model.add(Dense(units=1, activation='sigmoid'))
#this will have one output (units=1) predict 1/0 y/n the client has learnt
#sigmoid: takes op of previous layer and this layer and converts it to a value between 1/0 1=y/0=n 

```

```python
#compile model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
#optimizer = closest path to reduce loss
```

3. Fit,predict and evaluate

```python
model.fit(X_train,y_train, epochs=200, batch_size=32)
#epochs=how long we have to train(i.e higher epochs)will often result in a more accurate model but might lead to overfitting
#see that our loss is reducing and ideally your accuracy is increasing
```

```python
#prediction
y_hat =  model.predict(X_test)
#we store this in a var
#when we get our result from tensorflow its going to be a continuos val between 0/1
#convert it to binary outcome 0/1
#0 if val is less than 0.5 else 1

y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test,y_hat)
#accuracy isn't that great
#try training for 500 epochs, you should see the accuracy increase to 0.8 on this dataset and model setup
```

4. Saving and Reloading

```python
model.save('tfmodel.keras')
```
